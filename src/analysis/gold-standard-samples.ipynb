```python
tweet_files = defaultdict(dict)
```

```python
for p in path_e:
    for file in p.iterdir():
        if 'tweets' in file.stem:
            verb = twitter_data.extract_verb_from_filename(file)
            if verb is None:
                raise ValueError(f'Filename doesn\'t have a verb: {file.name}')
            
            if verb not in tweet_files[p.name].keys():
                tweet_files[p.name][verb] = set()
                
            tweet_files[p.name][verb].add(file)
```

```python
ids_dir = files.get_project_root()/'data'/'ids'/'twitter'
ids_dir.is_dir()
```

```python
ids = defaultdict(set)
```

```python
for d, verbs in tweet_files.items():
    dir_path = files.make_dir(ids_dir, d)
    
    for v, paths in verbs.items():
        v_path = files.make_dir(dir_path, v)
        
        for p in paths:
            tweet_ids = set(
                Tweets.from_csv(p, 'es', subset=['tweet_id']).d['tweet_id'].unique()
            )
            ids[v].update(tweet_ids)
```

```python
ids
```

```python
seed = 10000

sample_ids = defaultdict(set)
```

```python
for i, (v, d) in enumerate(ids.items()):
    rng = np.random.default_rng(seed + i)
    
    size = int(np.round(len(d)*0.1))
    s = rng.choice(list(d), size=size if size > 30 else 30, replace=False)
    
    sample_ids[v].update(set(s))
```

```python
# verify sample sizes
orig_len = {v: len(d) for v,d in ids.items()}
sample_len = {v: len(d) for v,d in sample_ids.items()}
```

```python
prob_verbs = ['ojala', 'decir', 'suplicar']
```

```python
path_save = files.get_project_root()/'data'/'gold-standard'/'2021-11-07-at-22:10:00'
path_save.is_dir()
```

```python
col_order = ['tweet_id', 'text_orig', 'created_at', 'author_id', 'tweet_place_id', 'referenced_tweets', 'lang', 'mentions', 'public_metrics.retweet_count', 'public_metrics.reply_count', 'public_metrics.like_count', 'public_metrics.quote_count']
```

```python
for v, paths in tweet_files['2021-11-07-at-22:10:00'].items():
    d_ids = sample_ids[v]
    data = pd.DataFrame()
    
    for i, path in enumerate(paths):
        d = pd.read_csv(path, usecols=col_order, sep='~', lineterminator='\n')
        x = d['tweet_id'].isin(d_ids)
        d = d.loc[x, :]
        
        if i==0:
            data = d
        else:
            data = pd.concat([data, d], axis=0)
        
    name = f'2021-11-07-gold-standard-tweets-{v}-{data.shape[0]}.xlsx'
    data = data.loc[:, col_order]
    data.to_excel(path_save/name, index=False)
```
